```python
# Function to compute min and max text length

import pandas as pd

# Function to compute min and max text length for a dataset split
def min_max_text_length(data, split):
    if split not in data:
        return None, None  # Handle missing splits

    df = pd.DataFrame(data[split])

    # Identify the correct text column dynamically
    text_column = next((col for col in ["text", "content", "description"] if col in df.columns), None)

    if text_column:
        df[text_column] = df[text_column].astype(str)  # Ensure all values are strings
        text_lengths = df[text_column].apply(len)  # Compute text lengths
        return text_lengths.min(), text_lengths.max()
    else:
        return None, None  # No valid text column found

# Compute min/max length for each dataset split
splits = ["train", "val", "test"]
length_results = {split: min_max_text_length(data, split) for split in splits}

# Print results
for split, lengths in length_results.items():
    print(f"{split.capitalize()} Set -> Min Length: {lengths[0]} characters, Max Length: {lengths[1]} characters")
```

```python
import spacy
from collections import defaultdict, Counter
nlp = spacy.load("en_core_web_sm")  # Ensure you have installed this: `python -m spacy download en_core_web_sm`

# Function to get top 5 tokens per class
# Tokenization function (Updated to remove numbers)
def text_pipeline_spacy(text):
    tokens = []
    doc = nlp(text)
    for t in doc:
        if not t.is_stop and not t.is_punct and not t.is_space and t.is_alpha:  # Exclude numbers
            tokens.append(t.lemma_.lower())  # Lemmatization to lowercase
    return tokens

def get_top_class_tokens(df):
    class_counters = defaultdict(Counter)

    for _, row in df.iterrows():
        label = row["label"]
        text_column = next((col for col in ["text", "content", "description"] if col in df.columns), None)
        if text_column:
            content = str(row[text_column])  # Ensure text is string format
            tokens = text_pipeline_spacy(content)
            class_counters[label].update(tokens)

    # Get top 5 tokens per class
    top_class_tokens = {label: counter.most_common(5) for label, counter in class_counters.items()}
    return top_class_tokens

# Compute top tokens per class from dataset_train
top_tokens_per_class = get_top_class_tokens(dataset_train)

# Print results
for label, tokens in top_tokens_per_class.items():
    print(f"Class {label}: {tokens}")
```

```python
import json
import pandas as pd
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Load dataset_train (Already cleaned)
# If dataset_train is already in memory, no need to load it again

dataset_train["label"] = dataset_train["label"]
ground_truth = dataset_train.set_index("id")["label"].to_dict()

# File paths for LLM-generated predictions (update if necessary)
llm_prompt_1_path = "llm_prompt_template_1.json"
llm_prompt_2_path = "llm_prompt_template_2.json"
llm_prompt_3_path = "llm_prompt_template_3.json"

# Function to load LLM predictions
def load_llm_predictions(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    predictions = {}
    for entry in data:
        try:
            # Convert predictions to integer, assign invalid values to class 5
            pred_label = int(entry["next_token"])
        except (ValueError, TypeError):
            pred_label = 5  # Assign invalid outputs to class 5

        predictions[entry["id"]] = pred_label


    return predictions

# Load predictions from the three prompt templates
llm_preds_1 = load_llm_predictions(llm_prompt_1_path)
llm_preds_2 = load_llm_predictions(llm_prompt_2_path)
llm_preds_3 = load_llm_predictions(llm_prompt_3_path)

# Function to evaluate model predictions
def evaluate_predictions(predictions, ground_truth):
    true_labels = []
    pred_labels = []

    for record_id, true_label in ground_truth.items():
        if record_id in predictions:

            true_labels.append(true_label)
            pred_labels.append(predictions[record_id])

    # Convert to integers before using sklearn metrics
    true_labels = [int(label) for label in true_labels]
    pred_labels = [int(label) for label in pred_labels]

    # Compute accuracy, precision, recall, and F1-score (macro-averaged)
    accuracy = accuracy_score(true_labels, pred_labels)
    precision, recall, f1, _ = precision_recall_fscore_support(true_labels, pred_labels, average="macro")

    return {
        "Accuracy": accuracy,
        "Macro Precision": precision,
        "Macro Recall": recall,
        "Macro F1-score": f1
    }

# Evaluate each LLM prompt template
results = {
    "Template 1": evaluate_predictions(llm_preds_1, ground_truth),
    "Template 2": evaluate_predictions(llm_preds_2, ground_truth),
    "Template 3": evaluate_predictions(llm_preds_3, ground_truth)
}

# Convert results to a DataFrame for better visualization
results_df = pd.DataFrame.from_dict(results, orient="index")

# Display results
print("LLM Prompt Template Evaluation Results:")
print(results_df)

# Save results to CSV (optional)
results_df.to_csv("llm_prompt_evafirluation_results.csv", index=True)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    logging_strategy = "epoch",
    output_dir="my_awesome_model", # HuggingFace wants a name for your model
    eval_strategy="epoch", # How often we want to evaluate the model
    learning_rate=learning_rate, # Hyperparameter
    per_device_train_batch_size=batch_size, # Hyperparameter
    per_device_eval_batch_size=batch_size, # Hyperparameter
    num_train_epochs=epochs, # Hyperparameter
    weight_decay=weight_decay, # Hyperparameter
    report_to='none' # Tells it to not use a tracking/reporting system (e.g. Weights & Biases)
)
```

```python
# Ensure labels are integers
train_dataset = train_dataset.map(lambda x: {"labels": int(x["labels"])})
```

```python
# Define tokenization function
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

# Apply tokenization
train_dataset = train_dataset.map(tokenize_function, batched=True)
validation_dataset = validation_dataset.map(tokenize_function, batched=True)


```

```python
from transformers import DataCollatorWithPadding

# Data Collator for dynamic padding
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Dictionary to store best models
best_models = {}

# Loop through base models and train each one
for model_name in base_models:
    print(f"Training model: {model_name}")

    # Load pre-trained model for classification
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=5)

    # Define Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=validation_dataset,
        tokenizer=tokenizer,
        data_collator=data_collator
    )

    # Train model
    trainer.train()

    # Save best model
    best_model_path = f"./best_model_{model_name.replace('/', '_')}"
    trainer.save_model(best_model_path)
    best_models[model_name] = best_model_path

print("Training complete! Best models saved.")
```
